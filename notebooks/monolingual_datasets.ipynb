```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_palette('Set3')
sns.set_context('talk')
```

```python
data = pd.read_csv(
    "../data/raw/monolingual_datasets.csv",
    sep="\t",
    dtype={"Language": str, "ISO code": str, "Datasets": 'Int64', "Models": 'Int64'},
    na_values=["-"],
)
```

```python
# top 10 datasets by size
datasets = data[:10].Datasets.tolist()
labels = data[:10]["ISO code"].tolist()

# datasets.append(data[10:].Datasets.sum())
# labels.append("other")

plt.figure(figsize=(12, 12))
plt.pie(datasets, labels=labels, labeldistance=0.9, autopct="%1.1f%%", pctdistance=0.7)
plt.savefig(
    "/Users/thesofakillers/Desktop/hf_datasets.png",
    dpi=300,
    bbox_inches="tight",
    pad_inches=0,
)
plt.show()
```
